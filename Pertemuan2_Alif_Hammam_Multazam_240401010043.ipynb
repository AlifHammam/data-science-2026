## Memuat Dataset Titanic

```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

print(df.head())
print(df.shape)
```

# Jawaban Latihan Q1 - Q6

## Q1. Jumlah baris, kolom, dan daftar kolom

```python
print("Shape:", df.shape)
print("\nKolom:")
print(df.columns.tolist())
```

## Q2. Missing Values

```python
missing = pd.DataFrame({
    'Jumlah Missing': df.isnull().sum(),
    'Persentase (%)': round(df.isnull().mean()*100,2)
})
missing[missing['Jumlah Missing']>0]
```

## Q3. Persentase Penumpang Selamat

```python
survival_rate = df['Survived'].mean()*100
print(f'Persentase selamat: {survival_rate:.2f}%')
```

## Q4. Wanita Kelas 1

```python
wanita_kelas1 = df[(df['Sex']=='female') & (df['Pclass']==1)]

print("Jumlah:", len(wanita_kelas1))
print("Rata-rata usia:", wanita_kelas1['Age'].mean())
```

## Q5. Tingkat Keselamatan per Kelas

```python
keselamatan = df.groupby('Pclass')['Survived'].mean().sort_values(ascending=False)
print(keselamatan)

print("\nKelas dengan tingkat keselamatan tertinggi:", keselamatan.index[0])
```

## Q6. Analisis Tambahan

```python
analisis = df.groupby('Sex')['Survived'].mean()*100
print(analisis)

print("\nInterpretasi:")
print("Penumpang perempuan memiliki tingkat keselamatan lebih tinggi dibanding laki-laki.")
```
