```python
# Import library
import pandas as pd
import numpy as np
```

```python
# Membaca dataset Titanic

link_data = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

titanic = pd.read_csv(link_data)

# Menampilkan 5 data pertama
titanic.head()
```

Dataset Titanic berisi informasi penumpang kapal Titanic seperti nama, usia, jenis kelamin, kelas penumpang, hingga status keselamatan.

```python
# Q1 - Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?

print("Ukuran dataset :", titanic.shape)

print("\nDaftar kolom:")
print(titanic.columns.tolist())
```

Dataset memiliki 891 baris dan 12 kolom.  
Setiap kolom menyimpan informasi berbeda mengenai penumpang Titanic.

```python
# Q2 - Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?

jumlah_kosong = titanic.isna().sum()

persentase_kosong = (jumlah_kosong / len(titanic)) * 100

ringkasan_missing = pd.DataFrame({
    "Jumlah Data Kosong": jumlah_kosong,
    "Persentase (%)": persentase_kosong
})

# Menampilkan hanya kolom yang memiliki missing value
ringkasan_missing[ringkasan_missing["Jumlah Data Kosong"] > 0]
```

Beberapa kolom masih memiliki data kosong, terutama pada kolom Cabin dan Age.

```python
# Q3 - Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy atau Pandas

total_selamat = titanic["Survived"].sum()

jumlah_penumpang = len(titanic)

persen_selamat = (total_selamat / jumlah_penumpang) * 100

print(f"Jumlah penumpang selamat : {total_selamat}")
print(f"Persentase keselamatan : {persen_selamat:.2f}%")
```

Hasil perhitungan menunjukkan bahwa kurang dari setengah penumpang Titanic berhasil selamat.

```python
# Q4 - Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa ratarata usianya?

female_first_class = titanic[
    (titanic["Sex"] == "female") &
    (titanic["Pclass"] == 1)
]

print("Total penumpang perempuan kelas 1 :", len(female_first_class))

print("Rata-rata usia :", round(female_first_class["Age"].mean(), 2), "tahun")
```

Penumpang perempuan pada kelas 1 cenderung memiliki usia dewasa.

```python
# Q5 - Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?

keselamatan_kelas = titanic.groupby("Pclass")["Survived"].mean() * 100

print(keselamatan_kelas)

kelas_teraman = keselamatan_kelas.idxmax()

print(f"\nKelas dengan tingkat keselamatan tertinggi adalah kelas {kelas_teraman}")
```

Penumpang kelas 1 memiliki peluang bertahan hidup paling tinggi dibanding kelas lainnya.

```python
# Q6 - Buat analisis tambahan berdasarkan variabel lain yang Anda anggap menarik. Sertakan interpretasi singkat.

keselamatan_gender = titanic.groupby("Sex")["Survived"].mean() * 100

print(keselamatan_gender)

print("\nAnalisis tambahan berdasarkan usia")

titanic["Kategori_Umur"] = titanic["Age"].apply(
    lambda x: "Anak-anak" if x < 18 else "Dewasa"
)

hasil_umur = titanic.groupby("Kategori_Umur")["Survived"].mean() * 100

print(hasil_umur)
```

Dari hasil analisis terlihat bahwa perempuan dan anak-anak memiliki tingkat keselamatan lebih tinggi.  
Hal ini berkaitan dengan prioritas evakuasi pada saat tragedi Titanic terjadi.
