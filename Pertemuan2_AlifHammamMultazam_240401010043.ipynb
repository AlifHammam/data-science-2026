## 2. NumPy: Komputasi Numerik

```python
import numpy as np

# Membuat array
a1D = np.array([1, 2, 3, 4, 5])
a2D = np.array([[1,2,3],[4,5,6]])

zeros = np.zeros((3, 4))
ones  = np.ones((2, 3))
eye   = np.eye(4)

arange = np.arange(0, 20, 2)
linsp  = np.linspace(0, 1, 11)

rng      = np.random.default_rng(seed=42)
rand_int = rng.integers(0, 100, (3, 4))
rand_flt = rng.normal(loc=0, scale=1, size=(100,))

print('a1D:', a1D)
print('a2D shape:', a2D.shape)
print('arange:', arange)
print('linspace:', linsp)
```

```python
# Vektorisasi dan Fungsi Statistik
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

print(a + b)
print(a * b)
print(a ** 2)

data = rng.normal(0, 1, 1000)
print(f'Mean : {np.mean(data):.4f}')
print(f'Std  : {np.std(data):.4f}')
print(f'Min  : {np.min(data):.4f}')
print(f'Max  : {np.max(data):.4f}')
print(f'Q1   : {np.percentile(data, 25):.4f}')
print(f'Q3   : {np.percentile(data, 75):.4f}')
```

## 3. Pandas: Manipulasi DataFrame

```python
import pandas as pd

# Pandas Series
nilai = pd.Series(
    [85, 92, 78, 95],
    index=['Alice', 'Bob', 'Caca', 'Dina']
)

print(nilai['Bob'])
print(nilai[nilai > 80])
print(nilai.mean())
print(nilai.describe())
```

```python
# Pandas DataFrame
data = {
    'nama': ['Ali', 'Budi', 'Caca', 'Dina'],
    'usia': [22, 25, 21, 23],
    'kota': ['Jakarta', 'Bandung', 'Surabaya', 'Medan'],
    'ipk' : [3.8, 3.5, 3.9, 3.7]
}
df = pd.DataFrame(data)

print(df.shape)
print(df.dtypes)
print(df)
```

## 4. Pertanyaan Latihan — Dataset Titanic

```python
# Load dataset Titanic
titanic_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(titanic_url)
df.head()
```

### Q1 — Jumlah baris, kolom, dan nama kolom

```python
print('Shape:', df.shape)  # (baris, kolom)
print('Kolom:', df.columns.tolist())
```

### Q2 — Missing values per kolom

```python
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)

missing_df = pd.DataFrame({'Jumlah Missing': missing, 'Proporsi (%)': missing_pct})
print(missing_df[missing_df['Jumlah Missing'] > 0])
```

### Q3 — Persentase penumpang yang selamat

```python
pct_selamat = np.mean(df['Survived']) * 100
print(f'Persentase selamat: {pct_selamat:.2f}%')
```

### Q4 — Filter penumpang wanita kelas 1

```python
wanita_k1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
print('Jumlah:', len(wanita_k1))
print(f'Rata-rata usia: {wanita_k1["Age"].mean():.2f} tahun')
```

### Q5 — Tingkat keselamatan per kelas penumpang

```python
survival_rate = df.groupby('Pclass')['Survived'].mean().round(3)
print(survival_rate)
print(f'\nKelas dengan tingkat keselamatan tertinggi: Kelas {survival_rate.idxmax()}')
```

### Q6 — Analisis tambahan: Tingkat keselamatan berdasarkan jenis kelamin

```python
# Saya tertarik melihat perbedaan tingkat selamat antara laki-laki dan perempuan
survival_sex = df.groupby('Sex')['Survived'].mean().round(3)
print(survival_sex)

# Breakdown lebih detail: kelas + jenis kelamin
detail = df.groupby(['Pclass', 'Sex'])['Survived'].mean().unstack().round(3)
print('\nDetail per kelas dan jenis kelamin:')
print(detail)
```
